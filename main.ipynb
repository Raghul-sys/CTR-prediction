```python
import pandas as pd
data = pd.read_csv('Statelevelip.csv')
```

```python
import plotly.express as px
px.bar(data,x='State',y='imp')
data['State'].nunique()
```

```python
data = data[data.State != 'None']
```

```python
data.isnull().sum()
```

```python
data=data.dropna(subset=['State'])
data['State'].nunique()
data['adtype'].nunique()
```

```python
data=data.dropna(subset=['adtype'])
data=data[data.adtype != 'None']
data=data[data.adtype != '4']
data=data[data.adtype != '0']
data=data[data.adtype != '59.0.3071.125']
data=data[data.adtype != '66.0.3359.181']
```

```python
data['exc'].nunique()
```

```python
data=pd.get_dummies(data,columns=['State'])
data=pd.get_dummies(data,columns=['exc'])
data=pd.get_dummies(data,columns=['adtype'])
```

```python
data['adtype']= data['adtype'].astype(int)
```

```python
x=data.drop(columns=['click','CTR','IP','imp'])
y=data['click']
type(x)
```

```python
type(x)
```

```python

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
data.corr()
```

```python
y_pred = regressor.predict(x_test)
```

```python
import sklearn.metrics as sm
print("R2 score =", round(sm.r2_score(y_test, y_pred), 2))
```
